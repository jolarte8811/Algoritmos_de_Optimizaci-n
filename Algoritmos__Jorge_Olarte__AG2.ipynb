```python
import math
import matplotlib.pyplot as plt
import numpy as np
import random

f = lambda x: x[0]**2 + x[1]**2
df = lambda x: [2*x[0] , 2*x[1]]
tasa = 0.2

iteraciones = 21
inicial = [10,16]

for i in range (iteraciones):
  a,b = df(inicial)
  inicial[0],inicial[1] = inicial[0]-(a*tasa),inicial[1]-(b*tasa)

inicial
```

```python
from numpy.core.function_base import linspace
resolucion=100
rango = 2.5
X = linspace(-rango,rango,resolucion)
Y = linspace(-rango,rango,resolucion)
Z = np.zeros((resolucion,resolucion))

for ix,x in enumerate(X):
  for iy, y in enumerate(Y):
    Z[iy,ix] = f([x,y])

count=plt.contourf(X,Y,Z,resolucion)
plt.colorbar(count)

#Generar punto de partida aleatorio
p = [random.uniform(-rango,rango),random.uniform(-rango,rango)]
plt.plot(p[0],p[1],"o",c="red")
TA=0.05

for _ in range(50):
  grad = df(p)
  p[0],p[1] = p[0]-TA*grad[0],p[1]-TA*grad[1]
  plt.plot(p[0],p[1],"o",c="red")

plt.plot(p[0],p[1],"o",c="white")
```
